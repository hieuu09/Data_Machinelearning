```python
# ham dinh duong
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cereal_df = pd.read_csv('https://raw.githubusercontent.com/CodexploreRepo/data-science/main/Code/A6_Seaborn/data/cereal.csv', sep=',')
cereal_df.head()
```

```python
fields=['shelf', 'weight', 'cups', 'rating']
cereal_df_new= cereal_df.drop(fields, axis=1)
cereal_df_new.head() # xoa casc cot ko can thiet
```

```python
# corr() nhan cac cot nao numeric de thuc hien cac heps tin
```

```python
cereal_corr=cereal_df_new.corr()
cereal_corr
# ma tran ding duong tuong quan voi nhau
# moi tuong quan
```

```python
# ve tuong quan voi nhau headmat
ones_corr= np.ones_like(cereal_corr, dtype=bool)
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape # kich thuoc cua hinh
```

```python
# ham triu return 1 nua duong true, va fale
mask= np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=cereal_corr, mask=mask); #mask lay 1 nua heatmask
adjusted_mask= mask[1:, :-1] # xoa hang so 1 va , kong lay hang dau tien, ko lay cot cuoi cung
adjusted_mask
```

```python
# ko lay hang dau tien , va ko lay cot cuoi cung
adjusted_cereal_corr= cereal_corr.iloc[1:, :-1]
```

```python

# su dung thang mau tu chon
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)
fig, ax= plt.subplots(figsize=(10, 8))
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,
            annot=True, annot_kws={"fontsize":13}, fmt=".2f", cmap=cmap, vmin=-1, vmax=1,
            linecolor='white', linewidths=0.5); # anot gia tri hien ben trong, lay  2 gia trij thap pha
# dieu chinh inhoa len
yticks = [i.upper() for i in adjusted_cereal_corr.index]
xticks = [i.upper() for i in adjusted_cereal_corr.columns]
ax.set_yticklabels(yticks, rotation=0, fontsize=13); # viet hoa tru y
ax.set_xticklabels(xticks, rotation=90, fontsize=13);
title='CORRELATION MATRIC'
ax.set_title(title, loc='left', fontsize=18);
```
