```python
# tru quan hoa du lieu californial
#truc quan hoa due lieu cac thanh pho
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('https://raw.githubusercontent.com/CodexploreRepo/data-science/main/Code/A4_Matplotlib/data/california_cities.csv', header =0, sep=',')
cities.head()
cities["area_total_km2"].sort_values(ascending=False)
```

```python
# xtract lat vi do and long kinh do
lat, lon = cities['latd'], cities['longd'] # lay kinh do va vi do
# lay danh so va dien tich
population, area= cities['population_total'], cities['area_total_sq_mi'] # cung cap danh so va dien tich
plt.style.use('seaborn') # lay style
plt.figure(figsize=(8, 6))
# do danh sao qua lon nen lay log 10 de thu nho , cmap
plt.scatter(lon, lat, c=np.log10(population), cmap='viridis',
            s=area, linewidths=0, alpha=0.5) # kinh do va vi do
  #chinh kich co s = area tung cham co area khac nhau, alpha do trong suot cua hinh

plt.xlabel('longlatitude');# gan cac title truc tung va truc hoanh
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}${population}') #hien thi thanh ba ben phai
plt.clim(3, 7) # gioi hanh mau tu 3 den 7


# create legen for cities size
plt.title('California Cities : opulatin and are distribution')
#plt.axis('equal')
# tao cac vong trong tuong ung tung loai dien tich, can sap xep lai dien tich ben tren theo  chiye

area_range=[50, 100, 300, 500]
for area in area_range:
  plt.scatter([], [], s=area, c='k', alpha=0.4,
              label=str(area)+'km$^2$')
  # khong can toa do nen de list rong
plt.legend(labelspacing=1, title='City Area') # thi moi show len, c='k' cho cung mau k
# khoang cach tung label
```
