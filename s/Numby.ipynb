```python
# xac xuat thong ke # do lech chuan , phuong sai, variance do phan tan
#standard do phan tan
# mean gia tri trung binh cong het chia count(sodo)
# variance do lech chuan inh phuon
dog_height=[600, 470, 170, 430, 300]
dog_he = np.array(dog_height)
print(dog_he)
```

```python
np.std(dog_he)
```

```python
# phuong sai
np.var(dog_he)
```

```python
# do lech chuan
np.sqrt(np.var(dog_he))
```
